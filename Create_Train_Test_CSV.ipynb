```python
training_real_dir = "/content/drive/My Drive/AM205_FinalProject/train/real"
training_fake_dir = "/content/drive/My Drive/AM205_FinalProject/fake"
#training_fake_dir = "/content/drive/My Drive/AM205_FinalProject/data_videos/manipulated_sequences/DeepFakeDetection/c40/images"

training_csv = "/content/drive/My Drive/AM205_FinalProject/train/train_data.csv"

testing_real_dir = "/content/drive/My Drive/AM205_FinalProject/test/real"
testing_fake_dir = "/content/drive/My Drive/AM205_FinalProject/fake"
testing_csv = "/content/drive/My Drive/AM205_FinalProject/test/test_data.csv"
```

```python
train_df = pd.DataFrame(columns = ['img1', 'img2', 'label'])
test_df = pd.DataFrame(columns = ['img1', 'img2', 'label'])
```

```python
# Train (real to real)
pbar = tqdm_notebook(total=len(os.listdir(training_real_dir)), desc="Train real to real")
for dir in os.listdir(training_real_dir):
    dirpath = dir
    for img1 in os.listdir(training_real_dir+ '/' + dirpath + '/'):
        img1_path = 'train/real/' + dirpath + '/' + img1
        for img2 in os.listdir(training_real_dir+ '/' + dirpath + '/'):
          img2_path = 'train/real/' + dirpath + '/' + img2
          train_df = train_df.append({'img1': img1_path, 'img2': img2_path, 'label': 0}, ignore_index = True)
    pbar.update(1)
pbar.close()
```

```python
# Train (real to fake)
pbar = tqdm_notebook(total=len(os.listdir(training_fake_dir)), desc="Train real to fake")
for dir in os.listdir(training_fake_dir):
    dirpath = dir

    pattern = re.compile(r'.*((0[1-9]|2[0-2])(__.*))__.*')
    result = pattern.match(dirpath)
    if result == None:
      continue
    original_folder = result.groups()[0]
    if os.path.isdir(training_real_dir+ '/' + original_folder + '/') == False:
      continue

    for img1 in os.listdir(training_fake_dir+ '/' + dirpath + '/'):
        img1_path = 'fake/' + dirpath + '/' + img1
        for img2 in os.listdir(training_real_dir+ '/' + original_folder + '/'):
          img2_path = 'train/real/' + original_folder + '/' + img2
          train_df = train_df.append({'img1': img2_path, 'img2': img1_path, 'label': 1}, ignore_index = True)
          break
    pbar.update(1)
pbar.close()
```

```python
# Test (real to real)
pbar = tqdm_notebook(total=len(os.listdir(testing_real_dir)), desc="Test real to real")
for dir in os.listdir(testing_real_dir):
    dirpath = dir
    for img1 in os.listdir(testing_real_dir+ '/' + dirpath + '/'):
        img1_path = 'test/real/' + dirpath + '/' + img1
        for img2 in os.listdir(testing_real_dir+ '/' + dirpath + '/'):
          img2_path = 'test/real/' + dirpath + '/' + img2
          test_df = test_df.append({'img1': img1_path, 'img2': img2_path, 'label': 0}, ignore_index = True)
    pbar.update(1)
pbar.close()
```

```python
# Test (real to fake)
pbar = tqdm_notebook(total=len(os.listdir(testing_fake_dir)), desc="Test real to fake")
for dir in os.listdir(testing_fake_dir):
    dirpath = dir

    pattern = re.compile(r'.*((2[3-8])(__.*))__.*')
    result = pattern.match(dirpath)
    if result == None:
      continue
    original_folder = result.groups()[0]
    if os.path.isdir(testing_real_dir+ '/' + original_folder + '/') == False:
      continue

    for img1 in os.listdir(testing_fake_dir+ '/' + dirpath + '/'):
        img1_path = 'fake/' + dirpath + '/' + img1
        for img2 in os.listdir(testing_real_dir+ '/' + original_folder + '/'):
          img2_path = 'test/real/' + original_folder + '/' + img2
          test_df = test_df.append({'img1': img2_path, 'img2': img1_path, 'label': 1}, ignore_index = True)
          break
    pbar.update(1)
pbar.close()
```
